```python
class Solution:
    def generateMatrix(self, n: int) -> List[List[int]]:
        # Initializing Matrix(which will be updated with required values)
        mat = [[0] * n for _ in range(n)]

        cur_row = cur_col = 0 # Initializing Current Row & Col
        
        # Initializing Boundaries
        top_boundary = 0
        bottom_boundary = n - 1
        left_boundary = 0
        right_boundary = n - 1

        # Key Variable
        direction = 'right' # Direction in which we have to fill the Val's, Initially in Right Direction

        for i in range(1, (n * n) + 1):
            # i --> Current Val, which need to be placed in Matrix
            mat[cur_row][cur_col] = i

            # As we placed One Val, Let's change the Row & Col numbers and Boundaries accordingly
            if direction == 'right': # Moving in Right Direction
                # Move the Column number(Same Row),, If it not reached Right Boundary
                if cur_col < right_boundary: # Cur_Col Should be Lesser than Right Bondary
                    cur_col += 1
                else: # cur_col >= right_boundary
                    direction = 'down' # Change Direction to "Down"
                    cur_row = cur_row + 1 # Move to Next Row
                    top_boundary = top_boundary + 1 # Update the Top Boundary
            
            elif direction == 'down': # Moving in Down Direction
                # Move the Row number(Same Column),, If it not reached Bottom Boundary
                if cur_row < bottom_boundary: # Cur_Row should be Lesser than Bottom Boundary
                    cur_row += 1
                else: # cur_row >= bottom_boundary
                    direction = 'left' # Change Direction to "Left"
                    cur_col = cur_col - 1 # Move to Left Column
                    right_boundary = right_boundary - 1 # Update the Right Boundary
            
            elif direction == 'left': # Moving in Left Direction
                # Move the Column number(Same Row),, If it not reached Left Boundary
                if cur_col > left_boundary: # Cur_Col should be Greater than Left Boundary
                    cur_col -= 1
                else: # cur_col >= left_boundary
                    direction = 'up' # Change Direction to "Up"
                    cur_row = cur_row - 1 # Move to Upward Row
                    bottom_boundary = bottom_boundary - 1 # Update the Bottom Boundary
            
            elif direction == 'up': # Moving in Up Direction
                # Move the Row number(Same Column),, If it not reached Top Boundary
                if cur_row > top_boundary: # Cur_Row should be Greater than Top Boundary
                    cur_row -= 1
                else: # cur_row >= top_boundary
                    direction = 'right' # Change Direction to "Right"
                    cur_col = cur_col + 1 # Move to Rightward Column
                    left_boundary = left_boundary + 1 # Update the Left Boundary

        return mat
```
